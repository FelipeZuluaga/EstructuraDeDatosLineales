  <h2 align="center">Estructuras de datos</h2>
  <h3 align="center">Tipos y aplicacion</h3>
  <h3>Estructura de datos estaticas:</h3>
<div>
  <p>De acuerdo con soyhenrry (2022) argumenta que las estructuras de datos estáticas son aquellas en las que el tamaño ocupado en memoria se define antes de que el programa se ejecute y no puede modificarse dicho tamaño durante la ejecución del programa, en ellas encontramos dos tipos:</p>
</div>

<table align="center">
  <tr>
    <th>Tipo</th>
    <th>Caracteristica</th>
  </tr>
  <tr>
    <td>Matrices</td>
    <td> 
      <p>Una matriz es una estructura de datos que permite almacenar un conjunto de datos del mismo tipo, con un único nombre se define la matriz y por medio de dos subíndices hacemos referencia a cada elemento de la misma</p>
    </td>
  </tr>
  <tr>
    <td>Arreglos</td>
    <td>
      <p>Los arreglos son estructuras de datos que viven en memoria principal de la computadora y la característica fundamental, de la memoria principal, es que el acceso a cualquier posición de memoria, es constante, ya sea para almacenar datos o recuperar datos de una posición de memoria</p>
    </td>
  </tr>
</table>



<h3>
  Estructura de datos dinamicas:
</h3>
<div style="text-align: justify">
  <p>Las estructuras de datos dinámicas son aquellas en las que el tamaño podrá modificarse durante la ejecución del programa; teóricamente no hay límites a su tamaño, salvo el que impone la memoria disponible en la computadora.</p>
</div>

<table align="center">
  <tr>
    <th>Tipo</th>
    <th>Caracteristicas</th>
  </tr>
  <tr>
    <td>lineales</td>
    <td> 
      <p>Las estructuras de datos lineales son aquellas en las que los elementos ocupan lugares sucesivos en la estructura y cada uno de ellos tiene un único sucesor y un único predecesor, es decir, sus elementos están ubicados uno al lado del otro relacionados en forma lineal.</p>
    </td>
  </tr>
  <tr>
    <td>No lineales</td>
    <td>
      <p>Las estructuras de datos no lineales, también llamadas multienlazadas, son aquellas en las que cada elemento puede estar enlazado a cualquier otro componente. Es decir, cada elemento puede tener varios sucesores o varios predecesores.
      </p>
    </td>
  </tr>
</table>





<div style="text-align: justify">
    <p>Dentro de los tipos de estructiras lineales y no lineales se encuentran otros tipos para cada uno de ellos:</p>
</div>

<table align="center">
  <tr>
    <th>lineales</th>
    <th>Caracteristicas</th>
    <th>No lineales</th>
    <th>Caracteristicas</th>
  </tr>
  <tr>
    <td>listas</td>
    <td> 
      <p>Una lista es una estructura dinámica de datos que contiene una colección 
      de elementos homogéneos (del mismo tipo) de manera que se establece entre ellos un orden.
      </p>
    </td>
    <td>
      <p>Arboles</p>
    </td>
    <td>
      <p>Un árbol es una estructura de datos ramificada (no lineal) que puede representarse como un conjunto de nodos enlazados entre sí por medio de ramas. La información contenida en un nodo puede ser de cualquier tipo simple o estructura de datos.</p>
    </td>
  </tr>
  <tr>
    <td>Pilas</td>
    <td>
      <p>Una pila o stack es una estructura de datos lineal en la que los datos son agregados y eliminados únicamente por un extremo de la estructura.</p>
    </td>
    <td>
      <p>Grafos</p>
    </td>
     <td>
      <p>Los grafos son estructuras de datos, es decir, tipos de datos abstractos, comúnmente los grafos son utilizados para el modelado de problemas. 
      Un grafo es un conjunto no vacıo de objetos o entes fısicos que tienen relación entre ellos.
      </p>
    </td>
  </tr>
  <tr>
    <td>Colas</td>
    <td>
      <p>
       Una cola es una estructura de datos que almacena elementos en una lista y permite acceder a los datos por uno de los dos extremos de la lista. 
       Un elemento se inserta en la cola (parte final) de la lista y se suprime o elimina por la frente (parte inicial, cabeza) de la lista.
      </p>
    </td>
    <td>
      <p></p>
    </td>
    <td>
      <p></p>
    </td>
  </tr>
</table>

<h2 align="center">CONCLUSIONES</h2>

<div style="text-align: justify,">
    <p>
        Es importante conocer los diversos tipos de datos estructurados que tenemos a nuestro alcance, comprender de sus caracteristicas y del como su aplicacion en el campo de la programacion resuelve problematicas muy puntuales y necesarios, para el caso de las estructuras de datos tipo pila y colas su implementacion en la industria y la tecnologia generan resultados optimos dada su seciencia de inicio y final (LIFO Y FIFO) dado que su implementacion en actividades industriales y automatizadas, eso sin mencionar las comerciales garantizan estandares altos de calidad y minimizando costos, es importante entonces reconocer que estas estructuras de informacion son una parte muy relevante para el desarrollo de tecnologuas o servicios acordes a estandares de calidad en la programacion.
